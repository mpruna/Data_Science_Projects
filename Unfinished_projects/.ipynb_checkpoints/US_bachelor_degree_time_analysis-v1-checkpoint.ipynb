### Exploring the dataset

[The Department of Education Statistics](https://nces.ed.gov/programs/digest/2013menu_tables.asp) releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category.

Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on [his personal website](http://www.randalolson.com/). You can download the dataset Randal compiled [here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv). Here's a preview of the first few rows:

```python
import pandas as pd
import numpy as np

# read dataset
bachelor_degrees=pd.read_csv('percent-bachelors-degrees-women-usa.csv')
bachelor_degrees.head()
```

As we can see each row represents an year and the women percentage in a particular bachelor degree. We get the man percentage by subtracting the women percentage on a given field out of 100.

We can group the bachelor degrees in the following categories:

   * stem (Science, technology, engineering, and mathematics)
   * liberal arts
   * other

### Compare the gender evolution

We will explore how gender percentage in bachelor degrees evolved from 1970 till 2012. We will do out comparison across stem, liberal arts and other categories.

To keep out code dry(don't repeat yourself) we will define a plotting function.

### Import plotly offline libraries

```python
from plotly import tools
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
from plotly.graph_objs import *
import plotly.graph_objs as go
#init_notebook_mode()
init_notebook_mode(connected=True)
config={'showLink': False,'displayModeBar': False}
#iplot(data, config={"displayModeBar": False})
import cufflinks as cf
cf.set_config_file(offline=True) #don't forget to set cufflinks offline
```

```python
# select columns & specify nr of plots
def time_analysis_plot(bachelor_categories,totalPolts,plt_col_nr):

    trace_m=[]
    trace_f=[]

    for i,j in zip(bachelor_categories,totalPolts):

        trace_f+=['trace_f'+str(i)]
        trace_f[j] = go.Scatter(x=bachelor_degrees['Year'],y=bachelor_degrees[i],line = dict(color = '#17BECF'),showlegend=False,
                opacity = 0.8)
        trace_m+=['trace_'+str(i)]
        trace_m[j] = go.Scatter(x=bachelor_degrees['Year'],y=100-bachelor_degrees[i],line = dict(color = '#7F7F7F'),showlegend=False,
                opacity = 0.8)
    
        if j==1:
            trace_f[j].update({'name': 'Women'})
            trace_f[j].update({'showlegend': True})
        
            trace_m[j].update({'name': 'Men'})
            trace_m[j].update({'showlegend': True})
        
    # create subplots titles 
    sp_titles=[str(item) for item in bachelor_categories]    

    #create figure
    fig = tools.make_subplots(rows=plt_col_nr, cols=1, subplot_titles=sp_titles,print_grid=False, shared_xaxes=True)

    for x,y,z in zip(trace_f,trace_m,range(1,7)):
        fig.append_trace(x,z,1)
        fig.append_trace(y,z,1)
    layout = go.Layout(showlegend=False)
    
    
    return fig
# Plot
#fig['layout'].update(height=1450,width=1000,autosize=True,margin=go.Margin(b=75))
#iplot(fig)
```

```python
# define bachelor categories

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

bachelor_categories=stem_cats
totalPlots=range(6)
plt_col_nr=6

#plot gender evolution in stem
fig=time_analysis_plot(bachelor_categories,totalPlots,plt_col_nr)
fig['layout'].update(height=1450,width=1000,autosize=True,margin=go.Margin(b=75))
iplot(fig,config=config)
```

```python
bachelor_categories=lib_arts_cats
totalPlots=range(5)
plt_col_nr=5

#plot gender evolution in stem
fig=time_analysis_plot(bachelor_categories,totalPlots,plt_col_nr)
fig['layout'].update(height=1450,width=1000,autosize=True,margin=go.Margin(b=75))
iplot(fig,config=config)
```

```python
config_cf={'showLink': False,'displayModeBar': False}

def plot_box_cf(df,title_plot,x_title,y_title):
    color='#f5f6f9'
    fig=df.iplot(kind='box',asFigure=True)
    
    layout = go.Layout(title=title_plot,plot_bgcolor=color,paper_bgcolor=color,xaxis=dict(title=x_title,showgrid=True,showticklabels=False),yaxis=dict(title=y_title,showticklabels=True))
    fig=go.Figure(fig,layout=layout)
    
    return fig
```

```python
df=bachelor_degrees.loc[:,stem_cats]
title_plot, x_title, y_title = 'Women distribution in stem degrees 1970-2011','Stem degrees','Percent'
fig=plot_box_cf(df,title_plot,x_title,y_title)
iplot(fig,config=config_cf)
```

```python
df=bachelor_degrees.loc[:,stem_cats]

def substract(x):
    r=100-x
    return r

df=df.apply(substract)

title_plot, x_title, y_title = 'Men distribution in stem degrees 1970-2011','Stem degrees','Percent'
fig=plot_box_cf(df,title_plot,x_title,y_title)
iplot(fig,config=config_cf)
```

```python
year=[1970,2011]

columns=stem_cats
df=bachelor_degrees

x1=df[df["Year"]==year[0]][columns].columns.tolist()
y1=df[df["Year"]==year[0]][columns].values.tolist()[0]

x2=df[df["Year"]==year[0]][columns].columns.tolist()
y2=df[df["Year"]==year[0]][columns].apply(lambda x: 100-x).values.tolist()[0]

x3=df[df["Year"]==year[1]][columns].columns.tolist()
y3=df[df["Year"]==year[1]][columns].values.tolist()[0]

x4=df[df["Year"]==year[1]][columns].columns.tolist()
y4=df[df["Year"]==year[1]][columns].apply(lambda x: 100-x).values.tolist()[0]

trace1=go.Bar(x=x1,
              y=y1,
              name='f 1970'
)

trace2=go.Bar(x=x2,
              y=y2,
              name='m 1970'
)

trace3=go.Bar(x=x3,
              y=y3,
              name='f 2011'
)

trace4=go.Bar(x=x4,
              y=y4,
              name='m 2011'
)


data = [trace1, trace2, trace3, trace4]
layout = go.Layout(barmode='stack')


fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='grouped-bar')
```
